# Exploratory data analysis top 200 YouTube-блогеров

In this project, the import and primary processing of data on the 200 largest YouTube channels (in terms of the number of views) was carried out, as well as a research analysis of the data with visualization of the results obtained.

<b>The aim of the project</b> is to study data on the 200 largest YouTube channels regarding their geography, the popularity of certain genres and categories of channels, the total number of likes and views, as well as the level of subscriber engagement.

<b>Data description:</b>

 - `Country`: Name of the country ('IN', 'US', 'KR', 'CA', 'BR', 'MX', 'SV', 'CL', 'NO', 'PR', 'BY', 'RU', 'PH', 'TH', 'AE', 'CO', 'ES', 'GB', 'AR', 'ID', 'NL','ES', 'IE', 'PK', 'AU', 'KW', 'SO')
 - `Channel Name`: Name of the channel
 - `Category`: Category ('Gaming & Apps','Sports', 'Music', 'Beauty & Fashion', 'Science & Tech', 'Fashion', 'LifeStyle')
 - `Main Video Category`: ('Music', 'Education', 'Shows', 'Gaming', 'Entertainment', 'People & Blogs', 'Sports', 'Howto & Style', 'Film & Animation', 'News & Politics', 'Pop music', 'Comedy', 'Nonprofits & Activism', 'Action-adventure game', 'Strategy video game', 'TV shows')
 - `Username`: Username of the youtube channel
 - `Followers`: No. of followers
 - `Main topic`: Topic that has been discussed more.
 - `More Topics`: Topics apart from main topic.
 - `Likes`: Total Likes
 - `Boost Index`: Boost index value.
 - `Engagement Rate`: Rate of engagement with the users.
 - `Engagement Rate 60days`: Rate of engagement with the users for 60 days
 - `View`: Total Views
 - `Views. Avg`: Average views
 - `Avg. 1 Day`: Average views for one day
 - `Avg. 3 Day`: Average views for 3 days
 - `Avg. 7 Day`: Average views for 7 days
 - `Avg. 14 Day`: Average views for 14 days
 - `Avg. 30 Day`: Average view for 30 days
 - `Avg. 60 Day`: Average view for 60 days
 - `Comments Avg`: Average comments.
 - `Youtube Link`: Link of the channel



## Data Preprocessing


### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
pd.set_option('display.max_columns', 22)
```

### Importing data
<b>Save data in the DataFrame and see the information about DataFrame.</b>

```python
folders = Path.cwd()
file_name = Path(folders, 'top_200_youtubers.csv')

df = pd.read_csv(file_name)
```

Finding size of the table:

```python
size = df.shape
display(size)
```

```python
df.info()
```

```python
df.head(10)
```

You can see that there are duplicates and missing null-values in the data.

### Removing duplicates

```python
df = df.drop_duplicates().reset_index(drop=True)
df.info()
df.shape
```

After removing duplicates size of the table has decreased. But you can see, that in the columns <b><i>Country, Category,Main vedio category, Main vedio topics, More topics, Engagement Rates,Views avg, Avg.1 Day,Avg. 3 Day, Avg. 7 Day, Avg. 14 Day, Avg. 30 Day, Comments Avg</i></b> has missing null-values.

### Handling missing values

Columns that lack only one value are of sufficient interest (such as Main topic, More topics, Engagement Rate, Comments Avg, etc.). Perhaps there is some one row in the data array with all the fields blank? Let's check it out:

```python
display(df[df['Main topic'].isna() == True])
```

Indeed, almost all fields in this row have no values. This kind of string is unrepresentative, so it can be deleted:

```python
df = df.drop(labels=75, axis=0)
```

<b>Find the fraction of the number of missing values:</b>

```python
count_col = df.shape[0]
round(((1-df.count()/count_col)*100),1).sort_values(ascending=False)
```

Among those fields in which there are no values, categorical variables can be distinguished, such as <b><i>Country, Category,</i></b> and quantitative, such as <b><i>Avg.1 Day,Avg. 3 Day, Avg. 7 Day, Avg. 14 Day, Avg. 30 Day</i></b>. To fill in the quantitative variables, we will use the average value, depending on the main topic of the video:

```python
df['Avg. 1 Day'] = df.groupby(['Main topic'])['Avg. 1 Day'].transform(lambda x : x.fillna(x.mean()))
df['Avg. 3 Day'] = df.groupby(['Main topic'])['Avg. 3 Day'].transform(lambda x : x.fillna(x.mean()))
df['Avg. 7 Day'] = df.groupby(['Main topic'])['Avg. 7 Day'].transform(lambda x : x.fillna(x.mean()))
df['Avg. 14 Day'] = df.groupby(['Main topic'])['Avg. 14 Day'].transform(lambda x : x.fillna(x.mean()))
df['Avg. 30 day'] = df.groupby(['Main topic'])['Avg. 30 day'].transform(lambda x : x.fillna(x.mean()))
df.info()
```

It can be seen that the columns <b><i>Avg.1 Day,Avg. 3 Day, Avg. 7 Day</i></b> have empty values. This is due to the fact that for some categories all the values were empty and the average could not be calculated. For them, we will use the average value for the entire field:

```python
df['Avg. 1 Day'] = df['Avg. 1 Day'].fillna(df['Avg. 1 Day'].mean())
df['Avg. 3 Day'] = df['Avg. 3 Day'].fillna(df['Avg. 3 Day'].mean())
df['Avg. 7 Day'] = df['Avg. 7 Day'].fillna(df['Avg. 7 Day'].mean())
df.describe()
```

<b>Let's look at rows of the dataset with missing Category:</b>

```python
display(df[df['Category'].isna() == True])
```

We can say with sufficient confidence that the channels that have <b><i>Music</i></b> or <b><i>Pop music</i></b> in the `Main Video Category` can be classified as <b><i>Music</i></b>. Similarly, <b><i>Gaming, Strategy video game</i></b> and <b><i>Action-adventure game</i></b> can be categorized as <b><i>Gaming & Apps</i></b>, and <b><i>Sports</i></b> to <b><i>Sports</i></b>. In addition, with a big stretch, <b><i>Education</i></b> can be attributed to <b><i>Science & Tech:</i></b>

```python
df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) &
            ((df['Main Video Category']=='Music') | 
             (df['Main Video Category']=='Pop music')),'Category'] = 'Music'

df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) &
           ((df['Main Video Category']=='Gaming') | 
            (df['Main Video Category']=='Strategy video game') | 
            (df['Main Video Category']=='Action-adventure game')),'Category'] = 'Gaming & Apps'

df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) & 
            (df['Main Video Category']=='Sports'),'Category'] = 'Sports'

df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) & 
           (df['Main Video Category']=='Education'),'Category'] = 'Science & Tech'

df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) & 
           ((df['Main Video Category']=='People & Blogs') | 
            (df['Main Video Category']=='Howto & Style')),'Category'] = 'LifeStyle'

```

However, there are still video categories such as <b><i>Shows, TV shows, Entertainment, Comedy,</i></b> etc. Therefore, I propose to introduce an additional category of channels - <b><i>Entertainment</i></b> and <b><i>Other:</i></b>

```python
df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) & 
           ((df['Main Video Category']=='Shows') | 
            (df['Main Video Category']=='TV shows') | 
            (df['Main Video Category']=='Entertainment') |
            (df['Main Video Category']=='Comedy') | 
            (df['Main Video Category']=='Film & Animation')),'Category'] = 'Entertainment'

df.loc[((df['Category'].isna() == True) | (df['Category']=='None')) & 
           ((df['Main Video Category']=='News & Politics') | 
            (df['Main Video Category']=='Nonprofits & Activism')),'Category'] = 'Other'
```

Unfortunately, we cannot fill in the gaps in the `Country` based on the current data set. We can't exclude rows with gaps in `Country` from the data set either, because they make up 18% of the total sample. Of course, we can fill them out based on the assumption that Maron 5, for example, being an American group and creating a YouTube channel, will indicate the USA as the country, and Shakira - Colombia, but since our goal is to conduct an initial research analysis, I think we can replace the empty values with <b><i>unfilled</i></b>. And if you need more in-depth research in the future, you can choose a more appropriate way to fill in the region.

```python
df['Country'] = df['Country'].fillna('unfilled')
df.info()
```

Great! All the values are filled in, so you can start visualizing the data.

## Analysis and visualization

```python
#sns.displot(df['Views'], kde=True,bins=20)
plt.figure(figsize=(17,7))

custom_params ={"lines.linewidth": 3, 
                "axes.spines.left": False,
                "axes.spines.bottom": False, 
                "axes.spines.right": False, 
                "axes.spines.top": False} 
sns.set(style="white", font='Segoe UI',font_scale=1.15,rc=custom_params)

#график по кол-ву каналов в странах
country = df.groupby('Country')['Channel Name'].count().sort_values(ascending=False).head(10)
sns.barplot(data=df, x=country.index, y=country, color='#CABBE9')
plt.xlabel('')
plt.ylabel('')
plt.yticks([])
plt.title('Number of channels from top 200 by country')
for i, val in enumerate(country.values):
    plt.text(i, val, val, horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':14})
plt.show()    

main_topic = df['Main topic'].value_counts().head(10)
category = df['Category'].value_counts()

plt.figure(figsize=(20,10))

#график топ категорий каналов
plt.subplot(1,2,1)
sns.barplot(data=df, x=category, y=category.index, color='#FFCEF3')
plt.title('Top categories by number of bloggers')
plt.xlabel('')
plt.xticks([])
for i, val in enumerate(category.values):
    plt.text(val+2, i+0.09, val, horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':14})
    
#график топ жанров
plt.subplot(1,2,2)
sns.barplot(data=df, x=main_topic, y=main_topic.index, color='#A1EAFB')
plt.subplots_adjust(wspace=0.3, hspace=0)
plt.title('Top genres by number of bloggers')
plt.xlabel('')
plt.xticks([])
for i, val in enumerate(main_topic.values):
    plt.text(val-1, i+0.14, val, horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':14})
    
plt.show()
```

Great, now we know which country has the most bloggers from the top 200 (US), and which categories of channels and genres are most often chosen by major bloggers. Let's see how categories and genres are distributed in the US:

```python
main_topic = df[df['Country']=='US']['Main topic'].value_counts().head(10)
category = df[df['Country']=='US']['Category'].value_counts()

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(data=df, x=category, y=category.index, color='#FFCEF3')
plt.title('Top Category in US')
plt.xlabel('')
plt.xticks([])
for i, val in enumerate(category.values):
    plt.text(val-1, i+0.09, val, horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':14})

plt.subplot(1,2,2)
sns.barplot(data=df, x=main_topic, y=main_topic.index, color='#A1EAFB')
plt.subplots_adjust(wspace=0.3, hspace=0)
plt.title('Top Genres in US')
plt.xlabel('')
plt.xticks([])
for i, val in enumerate(main_topic.values):
    plt.text(val-0.5, i+0.14, val, horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':14})
plt.show()
```

Next, let's see which channels have the highest number of likes/subscriptions/comments/engagement:

```python
more_like_channel = df.groupby('Channel Name')['Likes'].sum().nlargest(5)
more_followers_channel = df.groupby('Channel Name')['followers'].sum().nlargest(5)
more_views_channel = df.groupby('Channel Name')['Views'].sum().nlargest(5)
more_engr_channel = df.groupby('Channel Name')['Engagement Rate'].sum().nlargest(5)

sns.set(style="whitegrid", font='Segoe UI',font_scale=1.15,rc=custom_params)
plt.figure(figsize=(20,15))

#топ блогеров по лайкам
plt.subplot(2,2,1)
plt.subplots_adjust(wspace=0.3, hspace=0)
sns.barplot(data=df, x=more_like_channel, y=more_like_channel.index, color='#A1EAFB')
plt.title('Number of likes')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+09,0.50e+09,1.00e+09,1.50e+09,2.00e+09],labels = ['0 M','500 M','1 Bn','1.5 Bn','2 Bn'])

#топ блогеров по подписчикам
plt.subplot(2,2,2)
plt.subplots_adjust(wspace=0.5, hspace=0)
sns.barplot(data=df, x=more_followers_channel, y=more_followers_channel.index, color='#A1EAFB')
plt.title('Number of followers')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+08,0.50e+08,1.00e+08,1.50e+08,2.00e+08],labels = ['0 M','50 M','100 M','150 M','200 M'])

#топ блогеров по просмотрам
plt.subplot(2,2,3)
plt.subplots_adjust(wspace=0.3, hspace=0.2)
sns.barplot(data=df, x=more_views_channel, y=more_views_channel.index, color='#A1EAFB')
plt.title('Number of views')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+11,0.50e+11,1.00e+11,1.50e+11,2.00e+11],labels = ['0 M','50 Bn','100 Bn','150 Bn','200 Bn'])

#топ блогеров по вовлеченности
plt.subplot(2,2,4)
plt.subplots_adjust(wspace=0.4, hspace=0.2)
sns.barplot(data=df, x=more_engr_channel, y=more_engr_channel.index, color='#A1EAFB')
plt.title('Engagement rate')
plt.xlabel('')
plt.ylabel('')

plt.show()
```

From the graphs, it can be seen that the channel with the largest number of subscribers is T-Series, while it also leads in the number of views. It can be noted that 3 bloggers from the top 5 in terms of the number of subscribers got into the top 5 in terms of the number of views. Most likely, the number of views depends on the number of subscribers, let's check it out:

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(), cmap='cool', center=0, annot=True)
plt.show()
```

Indeed, the number of views has a fairly large correlation coefficient (0.83) with the number of subscribers. Let's see which genre of video gets the most likes:

```python
more_likes_genre = df.groupby('Main topic')['Likes'].sum().nlargest(5)
more_views_genre = df.groupby('Main topic')['Views'].sum().nlargest(5)

plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.subplots_adjust(wspace=0.3, hspace=0)
sns.barplot(data=df, x=more_likes_genre, y=more_likes_genre.index, color='#CABBE9')
plt.title('Genres with most likes')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+09,2.00e+09,4.00e+09,6.00e+09,8.00e+09],labels = ['0M','2 Bn','4 Bn','6 Bn','8 Bn'])

plt.subplot(1,2,2)
sns.barplot(data=df, x=more_views_genre, y=more_views_genre.index, color='#CABBE9')

plt.title('Genres with most views')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+11,2.00e+11,4.00e+11,6.00e+11,8.00e+11],labels = ['0 M','200 Bn','400 Bn','600 Bn','800 Bn'])

plt.show()
```

## Conclusions

During the analysis of the received dataset, we received the following results:

* <b>Categories and genres chosen by bloggers:</b> The most frequently encountered category among bloggers is Gaming & Apps (48%), and genres - Music (18%) and Entertainment (15%). If you look at the largest region (US), the categories do not change, and Lifestyle comes out on top in genres.
* <b>The largest blogger by number of followers</b> T-series (> 200 M followers)
* <b>The largest blogger by number of views</b> T-series (>150 Bn views)
* <b>The largest blogger by number of likes:</b> PewDiePie (>2 Bn likes)
* <b>The channel with the most engagement rate</b> Galinha Pintadinha
* <b>Genre with the most likes:</b> Lifestyle (>8 Bn likes)
* <b>Genre with the most views:</b> Music (>700 Bn views)

During the analysis, a hypothesis was put forward about the dependence of the number of views on the channel on the number of subscribers. Having calculated the correlation coefficient, we got 0.83, which confirms our hypothesis.
